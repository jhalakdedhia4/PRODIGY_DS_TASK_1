```python
# Sample DataFrame for demonstration (replace this with your actual DataFrame)
data = {
    'Country': ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France'],
    '2022': [331002651, 37742154, 212559417, 126476461, 83783942, 65273511]
}
df = pd.DataFrame(data)

# List of countries for the graph.
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

# Filtering the data frame
filtered_data = df[df['Country'].isin(countries_to_plot)]

# Creating the bar chart
plt.bar(filtered_data['Country'], filtered_data['2022'])
plt.xlabel('Country')
plt.ylabel('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right')  # Rotating the x-axis Labels for readability
plt.show()
```

```python
# Sample DataFrame for demonstration (replace this with your actual DataFrame)
data = {
    'Country Name': ['China', 'India', 'Russia', 'Japan', 'Canada', 'Australia', 'Germany', 'France', 'Samoa', 'South Korea'],
    '1998': [1262645000, 982223000, 146860000, 126770000, 30329000, 18160000, 82269000, 58700000, 17900, 46309000]
}
df = pd.DataFrame(data)

# Extracting required columns
countries = df['Country Name']
population_1998 = df['1998']

# List of countries
selected_countries = ['China', 'India', 'Russia', 'Japan', 'Canada', 'Australia', 'Germany', 'France', 'Samoa', 'South Korea']

# Filtering the DataFrame for the selected countries
selected_data = df[df['Country Name'].isin(selected_countries)]

# Sorting
sorted_data = selected_data.sort_values(by='1998', ascending=False)

# Creating bar chart
plt.figure(figsize=(11, 6))
plt.bar(sorted_data['Country Name'], sorted_data['1998'], color='pink')
plt.xlabel('Country')
plt.ylabel('Population in 1998')
plt.title('Distribution of Population of selected countries in 1998')
plt.xticks(rotation=45)
plt.show()
```
